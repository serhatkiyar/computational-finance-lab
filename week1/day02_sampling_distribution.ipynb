```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
n = 20
p = 0.3
```

```python
population_sigma = np.sqrt(n * p * (1 - p))
population_sigma
```

```python
sample_sizes = [10, 50, 100, 500, 1000, 2000, 5000]
```

```python
theoretical_se_list = []
simulated_se_list = []
```

```python
print(f"{'Sample Size(N)':<15} {'Theoretical SE':<15} | Simulation SE")

for size in sample_sizes:
    #Theoretical
    se_theory = population_sigma / np.sqrt(size)
    theoretical_se_list.append(se_theory)

    #Simulation
    num_trials = 1000
    samples = np.random.binomial(n, p, (num_trials, size))

    sample_means = samples.mean(axis=1)

    se_simule = sample_means.std()
    simule_hatalar = sample_means.std()

    simulated_se_list.append(se_simule)
    print(f"{size:<15} {se_theory:<15.6f} {se_simule:<15.6f}")


plt.figure(figsize=(10, 6))

plt.plot(sample_sizes, theoretical_se_list, 'r-', linewidth=3, label='Theoretical Formula (sigma / kok(n))')
plt.scatter(sample_sizes, simulated_se_list, color='blue', s=100, zorder=5, label='Simulated Data')

plt.title("Convergence of Standard Error as Sample Size Increases")
plt.xlabel("Sample Size (N)")
plt.ylabel("Standard Error (Volatility of the Mean)")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
