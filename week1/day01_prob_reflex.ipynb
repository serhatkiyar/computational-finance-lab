```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
data=np.array(np.random.choice(range(1,7), size=(10**5)))
data.shape
```

```python
data[:20]
```

```python
degerler, adetler = np.unique(data, return_counts=True)
```

```python
for d, a in zip(degerler, adetler):
    print(f"{d} : {a/10**5}")
```

```python
data = data.astype(np.float32)
```

```python
ortalamalar0 = np.zeros(100000, dtype=np.float32)
varyanslar0 = np.zeros(100000, dtype=np.float32)
for i in range(1, data.size + 1):
    ortalama = (data[:i]).mean()
    varyans = (np.sum(((data[:i] - ortalama)**2))) / i
    ortalamalar0[i - 1] = ortalama
    varyanslar0[i - 1] = varyans
```

```python
plt.figure(figsize=(10, 6))
limit = 1000
plt.plot(ortalamalar0[:limit], label="Ortalama (3.5' e yakinsar)", color='blue')
plt.plot(varyanslar0[:limit], label="Varyanslar (2.91' e yakinsar)", color='red')

plt.axhline(y=3.5, color='gray', linestyle='--', alpha=0.5, label="Teorik Ortalama ortalama")
plt.axhline(y=2.91, color='gray', linestyle='--', alpha=0.5, label="Teorik ortalama varyans")
plt.title("Buyuk Sayilar Yasasi: Ortalama ve Varyansin Kararliligi")
plt.xlabel("Atis Sayisi")
plt.ylabel("Degerler")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
ortalamalar = np.cumsum(data) / np.arange(1, data.size + 1)
ikinci_moment = np.cumsum(data**2) / np.arange(1, data.size + 1)
varyanslar = ikinci_moment - (ortalamalar**2)
```

```python
plt.figure(figsize=(10, 6))
limit = 1000
plt.plot(ortalamalar[:limit], label="Ortalama (3.5' e yakinsar)", color='blue')
plt.plot(varyanslar[:limit], label="Varyanslar (2.91' e yakinsar)", color='red')

plt.axhline(y=3.5, color='gray', linestyle='--', alpha=0.5, label="Teorik Ortalama ortalama")
plt.axhline(y=2.91, color='gray', linestyle='--', alpha=0.5, label="Teorik ortalama varyans")
plt.title("Buyuk Sayilar Yasasi: Ortalama ve Varyansin Kararliligi")
plt.xlabel("Atis Sayisi")
plt.ylabel("Degerler")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
