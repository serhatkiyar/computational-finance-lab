```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import seaborn as sns
```

```python
np.random.seed(42)
```

```python
n = 20
p = 0.3
trial_count = 10000 
```

```python
binom_sonuclari = np.random.binomial(n, p, trial_count)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(binom_sonuclari, bins=range(n+2), density=True, alpha=0.6, color='skyblue', edgecolor='black', label="Simulasyon (Binom)")
plt.title(f"Binom dagilim fonksiyonu (n={n}, p={p})")
plt.xlabel("basari sayisi")
plt.ylabel("deneme")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
mu = n * p
sigma = np.sqrt(n * p * (1 - p))
```

```python
print(mu, sigma)
```

```python
x_axis = np.linspace(0, n, 500)
y_axis = st.norm.pdf(x_axis, mu, sigma)
```

```python
plt.hist(binom_sonuclari, bins=range(n + 2), density=True, alpha=0.6, color='skyblue', edgecolor='black', label="Gercek Veri (Binom)")
plt.plot(x_axis, y_axis, 'r-', linewidth=3, label="Normal Yaklasm (Teori)")
plt.axvline(mu, color='k', linestyle='--', label=f"Ortalama ({mu})")
plt.title(f"Binom vs Normal Yaklasim")
plt.legend()
plt.show()
```

```python
sample_size = 1000
num_samples = 5000
```

```python
samples = np.random.binomial(n, p, (num_samples, sample_size))
sample_means = samples.mean(axis=1)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=50, density=True, color='orange', alpha=0.7, edgecolor="black", label="Orneklem Ortalamalari")
clt_sigma = sigma / np.sqrt(sample_size)
x_clt = np.linspace(mu -4*clt_sigma, mu +4*clt_sigma, 100)
y_clt = st.norm.pdf(x_clt, mu, clt_sigma)
plt.plot(x_clt, y_clt, "b--", linewidth=3, label="Teorik Normal (CLT)")
plt.title("Merkezi Limit Teoremi (CLT) Kaniti")
plt.xlabel("Ortalama Deger")
plt.legend()
plt.show()
```
