```python
import yfinance as yf
import os
```

```python
output_folder = 'data'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
    print(f"'{output_folder}' klasörü oluşturuldu.")

# 2. Veriyi Çek (SPY ETF - S&P 500)
# period='5y': Son 5 yıl
# interval='1d': Günlük mumlar
try:
    df = yf.download('SPY', period='5y', interval='1d', progress=False)


    if df.empty:
        print("veri bos")
    else:
        file_path = os.path.join(output_folder, 'SPY_5y_daily.csv')
        df.to_csv(file_path)
        
        print(f"kaydedildi {file_path}")
        print("-" * 30)
        print(f"Toplam Gün Sayısı: {len(df)}")
        print("İlk 5 Satır:")
        print(df.head())

except Exception as e:
    print("Hata {e}")
```
